```python
import json
with open('Character.json', encoding='utf-8') as f:
    char = json.load(f)
with open('Dynasty.json', encoding='utf-8') as f:
    dynasty = json.load(f)
with open('Event.json', encoding='utf-8') as f:
    event = json.load(f)
with open('Festival.json', encoding='utf-8') as f:
    festival = json.load(f)
with open('Place.json', encoding='utf-8') as f:
    place = json.load(f)
```

```python
for sample in char:
    sample['type'] = 'Character'
for sample in dynasty:
    sample['type'] = 'Dynasty'
for sample in event:
    sample['type'] = 'Event'
for sample in festival:
    sample['type'] = 'Festival'
for sample in place:
    sample['type'] = 'Place'
```

```python
full = char + dynasty + event + festival + place
```

```python

def write_jsonl(records, path):
    with open(path, "w", encoding="utf-8") as w:
        for r in records:
            w.write(json.dumps(r, ensure_ascii=False) + "\n")
```
